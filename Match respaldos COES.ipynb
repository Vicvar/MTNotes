```python
def getDate(file, dr):
    m = re.search('\d{4}-\d{1,2}-\d{1,2}',file)
    if m:
        dates[0][file]=m.group()
    else:
        m = re.search('\d{1,2}-\d{1,2}-\d{4}',file)
        if m:
            dates[1][file] = m.group()
        else:
            m = re.search('\d{1,2}-\d{1,2}-\d{2}',file)
            if m:
                dates[2][file] = dr+'-'+m.group()
            else:
                m = re.search('_\d{1,2}_\d{1,2}_\d{1,2}_',file)
                if m:
                    dates[3][file] = m.group()
                else:
                    m = re.search('\d{1,2}\s-\s\d{1,2}\s-\s\d{4}',file)
                    if m:
                        dates[4][file] = m.group()
                    else:
                        m = re.search('\d{2}\.\d{2}',file)
                        if m:
                            dates[5][file] = dr+"."+m.group()
                        else:
                            m = re.search('_\d{2}_\d{1,2}_\d{4}',file)
                            if m:
                                dates[6][file] = m.group()
                            else:
                                m = re.search('\d{4}_\d{2}_\d{2}_',file)
                                if m:
                                    dates[7][file] = m.group()
                                else:
                                    m = re.search('\d{1,2}_\d{2}_\d{2}_',file)
                                    if m:
                                        dates[8][file] = m.group()
                                    else:
                                        m = re.search('\d\s-\s\d{2}\s-\d{4}',file)
                                        if m:
                                            dates[9][file] = m.group()
                                        else:
                                            print('NO MATCH!: ',file)

dates = [{},{},{},{},{},{},{},{},{},{}]
print(os.listdir("./COES/Respal2"))

base_dir = "./COES/Respal2/1. Respaldos Noticias"
for dr in os.listdir(base_dir):
    if(dr.endswith(".xlsx")):
        continue
    else:
        for file in os.listdir(base_dir+"/"+dr):
            if file != 'Combinados':
                getDate(file,dr)
            else:
                for fc in os.listdir(base_dir+"/"+dr+"/Combinados"):
                    getDate(fc,dr)
                    
base_dir = "./COES/Respal2/2. Eventos restantes"

for dr in os.listdir(base_dir):
    if(dr.endswith(".xlsx")):
        continue
    else:
        if dr != 'Combinados':
            getDate(dr,'9999')
        else:
            for fc in os.listdir(base_dir+"/Combinados"):
                getDate(fc,'9999')            
```
