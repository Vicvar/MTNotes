```python
search = "http://seia.sea.gob.cl/busqueda/buscarProyectoAction.php"
xls = "http://seia.sea.gob.cl/busqueda/buscarProyectoActionExcel.php"
params = {"busca":"true",
"externo":"1",
"modo":"normal",
"nombre":"",
"presentacion":"AMBOS",
"fecha":"1",
"dia_desde":"1",
"mes_desde":"5",
"ano_desde":"1992",
"dia_hasta":"6",###
"mes_hasta":"5",###
"ano_hasta":"2019",###
"diac_desde":"1",
"mesc_desde":"5",
"anoc_desde":"1992",
"dia_hastac":"6",
"mes_hastac":"5",
"ano_hastac":"2019",
"estado":"",
"sector":"",
"id_tipoexpediente":""}
```

```python
s = requests.session()
s.post(url=search, data=params)
r = s.get(url=xls)
f = open("SEIA2/XLS/seia0.xls","wb")
f.write(r.content)
f.close()
```

```python
tot_df = pd.DataFrame()
part_df = pd.read_excel("SEIA2/XLS/seia0.xls")
part_df = part_df.drop(part_df.index[1000])

file = "SEIA2/XLS/seia"
date = part_df.loc[999,"Fecha presentación"]

params["fecha"]="1"

ini=time.time()

i=0
while len(part_df)>999:
    i+=1
    tot_df = tot_df.append(part_df)
    print(date)
    params["dia_hasta"] = str(date.day)
    params["mes_hasta"] = str(date.month)
    params["ano_hasta"] = str(date.year)
    
    s.post(url=search, data=params)
    r = s.get(url=xls)
    f = open("SEIA2/XLS/seia"+str(i)+".xls","wb")
    f.write(r.content)
    f.close()
    
    part_df = pd.read_excel("SEIA2/XLS/seia"+str(i)+".xls")
    
    if ((part_df.index==1000).any()):
        part_df = part_df.drop(part_df.index[1000])
        date = part_df.loc[999,"Fecha presentación"]

tot_df = tot_df.append(part_df)

print("Total:",len(tot_df))
print(time.time()-ini)
```

```python
newdf = tot_df.reset_index(drop=True)
```
