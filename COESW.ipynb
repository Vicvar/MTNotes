```python
#Cast dates to datetime

#fecha_observacion
fecha_observacion = acciones[['fecha_observacion_d','fecha_observacion_m','fecha_observacion_a']]
fecha_observacion.columns = ['day','month','year']
print(fecha_observacion.iloc[0:10])
fecha_observacion.day = pd.to_numeric(fecha_observacion.day,errors='coerce')
fecha_observacion.month = pd.to_numeric(fecha_observacion.month,errors='coerce')
fecha_observacion.year = pd.to_numeric(fecha_observacion.year,errors='coerce')
fecha_observacion.year = fecha_observacion.year + 2000
fecha_observacion = pd.to_datetime(fecha_observacion,errors='coerce')
print(fecha_observacion.iloc[0:10])

#fecha_inicio
fecha_inicio = acciones[['fecha_inicio_d','fecha_inicio_m','fecha_inicio_a']]
fecha_inicio.columns = ['day','month','year']
fecha_inicio.day = pd.to_numeric(fecha_inicio.day,errors='coerce')
fecha_inicio.month = pd.to_numeric(fecha_inicio.month,errors='coerce')
fecha_inicio.year = pd.to_numeric(fecha_inicio.year,errors='coerce')
fecha_inicio.year = fecha_inicio.year + 2000
fecha_inicio = pd.to_datetime(fecha_inicio,errors='coerce')

#fecha_fin
fecha_fin = acciones[['fecha_fin_d','fecha_fin_m','fecha_fin_a']]
fecha_fin.columns = ['day','month','year']
fecha_fin.day = pd.to_numeric(fecha_fin.day,errors='coerce')
fecha_fin.month = pd.to_numeric(fecha_fin.month,errors='coerce')
fecha_fin.year = pd.to_numeric(fecha_fin.year,errors='coerce')
fecha_fin = fecha_fin.dropna().astype(int)
fecha_fin.year = fecha_fin.year + 2000
fecha_fin = pd.to_datetime(fecha_fin,errors='coerce')
fecha_fin.head()
```

```python
acciones = acciones.drop(['fecha_observacion_d','fecha_observacion_m','fecha_observacion_a',
                          'fecha_inicio_d','fecha_inicio_m','fecha_inicio_a',
                          'fecha_fin_d','fecha_fin_m','fecha_fin_a'],axis=1)

acciones = pd.concat([acciones,fecha_observacion,fecha_inicio,fecha_fin],axis=1)
acciones = acciones.rename(columns={0:'fecha_observacion',1:'fecha_inicio',2:'fecha_fin'})
acciones.head()
```
