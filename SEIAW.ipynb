```python
import pandas as pd
import re
```

```python
original = pd.read_excel("SEIA2/seia_3_5_2019.xls")
original.dtypes
```

```python
original.head()
```

```python
original = original.drop(['Región','Provincias'], axis=1)
original.head()
```

```python
original = original.drop(original[original.Estado=='En Admisión'].index)
original.head()
```

```python
original = original.sort_values(by='Fecha presentación')
```

```python
original.reset_index(drop=True,inplace=True)
original.head()
#original.iloc[100:110]
```

```python
#comc = original[original['Comunas'].str.match(r'^[a-z A-Z]+-([a-z A-Z]+)?$',na=False)]
#comc = comc[~comc['Comunas'].str.contains('Til')]
#comc[['Comunas']].head(1000)
```

```python
comunas = []

for i in range(len(original)):
    try:
        split = original.iloc[i]['Comunas'].split('-')
    except:
        print(i,'NaN?')
    tilted = False
    for s in split:
        if (s=='Til'):
            if(not tilted):
                comunas.append([i,'TILTIL'])
                tilted=True
        else:
            comunas.append([i,s.upper()])

comunas_nuevas = pd.DataFrame(comunas, columns = ['id_proyecto', 'comuna'])
```

```python
comunas_nuevas.head()
```

```python
codigos_comuna = pd.read_csv("Localidades/comunas.csv")
codigos_comuna['Nombre Comuna'] = codigos_comuna['Nombre Comuna'].str.upper()
codigos_comuna.head()
```

```python
codigos_comuna = codigos_comuna.drop(['Código Provincia'], axis=1)
```

```python
final = comunas_nuevas.join(codigos_comuna.set_index('Nombre Comuna'),on='comuna')
final = final.rename(columns = {'comuna':'nombre_comuna','Código Comuna 2017':'comuna'})
#final['comuna'] = final['comuna'].astype(int) 
#final.loc[final.comuna.isnull()].drop_duplicates(subset='nombre_comuna')
final.head()
```

```python
#Reemplazar valores de comunas_nuevas por valores correspondientes a los codigos
comunas_fixed = comunas_nuevas.replace({'SAN JOSÉ DE LA MARIQUINA':'MARIQUINA',
                                        'SAN JOSÉ DE LA MARIQU':'MARIQUINA',
                                        'REQUINOA':'REQUÍNOA',
                                        'REQUI':'REQUÍNOA',
                                        'LA CALERA':'CALERA',
                                        'AYSÉN':'AISÉN',
                                        'COYHAIQUE':'COIHAIQUE',
                                        'COYHAI':'COIHAIQUE',
                                        'LLAY LLAY':'LLAILLAY',
                                        'PUERTO SAAVEDRA':'SAAVEDRA',
                                        'LAS GUAITECAS':'GUAITECAS',
                                        'TREHUACO':'TREGUACO',
                                        'VILLA OHIGGINS':'O\'HIGGINS'})
#Completar valores incompletos de ser posible
fix = final.loc[final.comuna.isnull()].drop_duplicates(subset='nombre_comuna')['nombre_comuna'].tolist()
rep = {}
for n in fix:
    lu =codigos_comuna.loc[codigos_comuna['Nombre Comuna'].str.match(re.escape(n)+'.*')]
    if len(lu)==1:
        rep[n]=lu.iloc[0]['Nombre Comuna']
comunas_fixed = comunas_fixed.replace(rep)
comunas_fixed.head()
```

```python
#Explorando autocompl
codigos_comuna.loc[codigos_comuna['Nombre Comuna'].str.contains('PICHI')]
```

```python
intermedio = comunas_fixed.join(codigos_comuna.set_index('Nombre Comuna'),on='comuna')
intermedio = intermedio.rename(columns = {'comuna':'nombre_comuna','Código Comuna 2017':'comuna'})
#nombres de comunas incompletables
incomp = intermedio.loc[intermedio.comuna.isnull()].drop_duplicates(subset='nombre_comuna')
print(incomp.head())
print(len(incomp))
```

```python
estesi = intermedio.drop(intermedio.loc[intermedio.comuna.isnull()].index)
```

```python
estesi.head()
```

```python
comunasEnProyectos = estesi.drop(['nombre_comuna'],axis=1)
comunasEnProyectos['comuna'] = comunasEnProyectos['comuna'].astype(int)
comunasEnProyectos.set_index('id_proyecto',drop=True, inplace=True)
comunasEnProyectos.head()
```

```python
proyectos = original.drop(['Comunas'],axis=1)[['Nombre','Tipo','Tipología','Titular','Inversión (MMU$)','Estado','Fecha presentación','Fecha calificación','Sector productivo','Latitud punto representativo','Longitud punto representativo']]
```

```python
comunas_ids = comunas_imfd[['id','cut']]
cids_dict = comunas_ids.to_dict()['cut']
cids_idict = dict((v,k) for k, v in cids_dict.items())
comunasEnProyectos.comuna = comunasEnProyectos.comuna.replace(cids_idict)
comunasEnProyectos = comunasEnProyectos.rename(columns={'comuna':'id_comuna'})
comunasEnProyectos['id_proyecto'] = comunasEnProyectos.index
comunasEnProyectos = comunasEnProyectos.reset_index(drop=True)
comunasEnProyectos = comunasEnProyectos[['id_proyecto','id_comuna']]
comunasEnProyectos.head()
```
