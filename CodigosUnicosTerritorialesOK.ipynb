```python
import pandas as pd
```

```python
todo = pd.read_excel("../CodigosUnicosTerritoriales_2018_v03.xls")
todo.head()
```

```python
regiones = todo[['Código Región','Nombre Región']]
regiones = regiones.drop_duplicates().set_index('Código Región')
regiones.head(20)
```

```python
provincias = todo[['Código Provincia','Nombre Provincia','Código Región']]
provincias = provincias.drop_duplicates().set_index('Código Provincia')
provincias.head(20)
```

```python
comunas = todo[['Código Comuna 2017','Nombre Comuna','Código Provincia']]
comunas = comunas.drop_duplicates()
comunas = comunas.set_index('Código Comuna 2017')
comunas.head(20)
```

```python
len(comunas)
```

```python
regiones.to_csv("Localidades/regiones.csv")
provincias.to_csv("Localidades/provincias.csv")
comunas.to_csv("Localidades/comunas.csv")
```
