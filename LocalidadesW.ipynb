```python
import pandas as pd
```

```python
total = pd.read_csv("Localidades/localidades.csv")
```

```python
total.head()
```

```python
regiones = total[['REGION','NOM_REGION']]
regiones = regiones.drop_duplicates()
regiones = regiones.set_index('REGION')
print(regiones)
```

```python
provincias = total[['PROVINCIA','NOM_PROVINCIA','REGION']]
provincias = provincias.drop_duplicates()
provincias = provincias.set_index('PROVINCIA')
provincias.head()
```

```python
comunas = total[['COMUNA','NOM_COMUNA','PROVINCIA']]
comunas = comunas.drop_duplicates()
comunas = comunas.set_index('COMUNA')
comunas.head()
```

```python
localidades = total[['NOM_LOC','longitude','latitude','COMUNA','DISTRITIO']]
localidades = localidades.rename(columns={'DISTRITIO':'DISTRITO_CENSAL'})
localidades = localidades.drop_duplicates()
localidades = localidades.reset_index(drop=True)
localidades.head()
```

```python
regiones.to_csv("Localidades/regiones.csv")
provincias.to_csv("Localidades/provincias.csv")
comunas.to_csv("Localidades/comunas.csv")
localidades.to_csv("Localidades/localidades_only.csv")
```
